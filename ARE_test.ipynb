```python
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
def person(fume):
    X = random.randint(0,100)
    Y = random.randint(0,100)
    return [X,Y,fume]
```

```python
def pop_init(nb,pourc):
    pop = []
    nbf = int(nb*pourc)
    for i in range(nbf):
        pop.append(person('red'))
    for i in range(nb-nbf):
        pop.append(person('white'))
        
    pop = np.array(pop)
    plt.scatter(pop[:,0],pop[:,1], c = pop[:,2], edgecolors = 'grey')
    ax = plt.gca()
    ax.axes.xaxis.set_visible(False)
    ax.axes.yaxis.set_visible(False)
    plt.show()
    
    return pop
    
```

```python
pop = pop_init(1000,0.3)
```
