```python
def interaction_proche(pop):
    proche = []
    x_val = [x[0] for x in pop]
    y_val = [x[1] for x in pop]
    for i in range(len(pop)):
        proche.append([])
    for i in range (len(pop)):
        dist = [np.sqrt((x_val[i] - x[0])**2 + (y_val[i] - x[1])**2) for x in pop]
        pgeo_i = [dist.index(j) for j in dist if j < 200]
        psoc_i = []
        for j in range(len(pgeo_i)):
            if i != j and j not in proche[i]:
                pi = pop[i]
                pj = pop[j]
                inter = random.random()
                if pi[5] == pj[5]: #même catégorie socio-professionnelle
                    if pi[3]-pj[3]<10:
                        if inter < 0.001:
                            proche[i].append(j)
                            proche[j].append(i)
                    elif pi[3]-pj[3]<20:
                        if inter < 0.0012:
                            proche[i].append(j)
                            proche[j].append(i)
                    else:
                        if inter < 0.0001:
                            proche[i].append(j)
                            proche[j].append(i)
                        
                else:
                    if pi[3]-pj[3]<10:
                        if inter < 0.0003:
                            proche[i].append(j)
                            proche[j].append(i)
                    elif pi[3]-pj[3]<20:
                        if inter < 0.00008:
                            proche[i].append(j)
                            proche[j].append(i)
                    else:
                        if inter < 0.00001:
                            proche[i].append(j)
                            proche[j].append(i)


    i=0
    while i<len(proche):
        for j in proche[i]:
            change=[]
            if pop[j][-1]:
                for a in proche[i]:
                    if pop[a][-1]==False :
                        change.append(a)
            fume = random.random()
            if fume<0.05 and len(change)>1:
                n=change[random.randint(0,len(change)-1)]
                if pop[n][-4]!=0:
                    pop[n][-1]=True
        i=i+1

    return pop

pop = create_pop(100)
print(compte(pop))
interaction_proche(pop)
compte(pop)

                 

        
```

```python
def compte(pop):
    r = 0
    for i in range(len(pop)):
        if pop[i][-1]:
            r = r+1
    return r
```

```python
def affiche(pop, date = 0):
    
    fig = plt.figure(figsize = (6,6))
    
    x_val = [x[0] for x in pop]
    y_val = [x[1] for x in pop]
    color = [x[2] for x in pop]
            
    plt.scatter(x_val,y_val, c = color, edgecolors = 'black')
    ax = plt.gca()
    ax.axes.xaxis.set_visible(False)
    ax.axes.yaxis.set_visible(False)
    
    year = 2022 + date
    plt.title(str(year))

    plt.xlim(0,100)
    plt.ylim(0,100)
    plt.show()
```

```python
def move_year(pop):
    for i in range(len(pop)):
        pop[i][0] += random.randint(-5,5) 
        if pop[i][0]<0:
            pop[i][0] = 0
        elif pop[i][0]>100:
            pop[i][0] = 100
            
        pop[i][1] += random.randint(-5,5)
        if pop[i][1]<0:
            pop[i][1] = 0
        elif pop[i][1]>100:
            pop[i][1] = 100
    interaction_proche(pop)        
    debute_age_sponta(pop)
    arret_reprise_sponta(pop)
    diffu_pub(pop)
    diffu_lutte(pop)
    
        
    return pop
```

```python
def simulation(pop,year):
    
    date = 0
    nbf = [sum(bool(x[10]) for x in pop)/len(pop)*100]

    for i in range(1,year+1):
        pop = move_year(pop)
        date += 1
        nbf.append(sum(bool(x[-1]) for x in pop)/len(pop)*100)
        x_val = [x[0] for x in pop]
        y_val = [x[1] for x in pop]
        color = [x[2] for x in pop]
    
    affiche(pop,date)

    plt.plot(nbf)
    axes = plt.gca()
    axes.set_xlabel('Date')
    axes.set_ylabel('% de fumeur')
    plt.ylim(0,100)
    plt.xlim(0,year)
    plt.show()
```
