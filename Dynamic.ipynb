```python
def proche(pop):
    proche = []
    for i in range (len(pop)):
        x_val = [x[0] for x in pop]
        y_val = [x[1] for x in pop]
        dist = [np.sqrt((x_val[i] - x[0])**2 + (y_val[i] - x[1])**2) for x in pop]
        p_i = [dist.index(j) for j in dist if j < 3]
        proche.append(p_i)
    return proche

pop = [[13, 12, 'white', 35, 'h', 'ouvrier', 'healthy', False], \
    [14, 12, 'white', 20, 'f', 'employe', 'healthy', False], \
    [14, 11, 'red', 25, 'h', 'chomeur', 'healthy', True], \
    [45, 60, 'white', 76, 'h', 'retraite', 'healthy', False], [56, 67, 'red', 72, 'h', 'retraite', 'healthy', True]]

print(pop)
print("\n")  
print(proche(pop))

i=0
tmp=[]
while i<len(proche(pop)) :
    if len(proche(pop)[i])>1 :
        if i not in tmp :
            for b in range(1,len(proche(pop)[i])) :
                tmp.append(b)
            for j in proche(pop)[i]:
                change=[]
                if pop[j][-1]:
                    for a in proche(pop)[i]:
                        if pop[a][-1]==False :
                            change.append(a)
            fume = random.random()
            if fume<0.3 :
                pop[math.floor(len(change)*random.random())][-1]=True
    i=i+1

print("\n")                
print(pop)


                 

        
```

```python
def arret_reprise_sponta(pop): #chaque année
    for i in range(len(pop)) :
        if pop[i][-4]<4 :
            n = random.random()
            if n<0.01 :
                pop[i][-4]=pop[i][-4]+1
                pop[i][-1]=False
                pop[i][2]='white'
                reprise = random.random()
                if reprise<0.7 :
                    pop[i][-1]=True
                    pop[i][2]='red'
                    
        
pop = create_pop(20)
print(pop)
arret_reprise_sponta(pop)
print(pop)
```

```python
def diffu_lutte(pop):
    for i in range(len(pop)) :
        if pop[i][-4]<4 :
            n = random.random()
            if n<0.01 :
                pop[i][-4]=pop[i][-4]+1
                pop[i][-1]=False
                pop[i][2]='white'
                reprise = random.random()
                if reprise<0.7 :
                    pop[i][-1]=True
                    pop[i][2]='red'
```

```python
def pop_init(nb):
    pop = create_pop(nb)
    debute_age_sponta(pop)
    arret_reprise_sponta(pop)
    diffu_lutte(pop)
    x_val = [x[0] for x in pop]
    y_val = [x[1] for x in pop]
    color = [x[2] for x in pop]
            
    plt.scatter(x_val,y_val, c = color, edgecolors = 'black')
    ax = plt.gca()
    ax.axes.xaxis.set_visible(False)
    ax.axes.yaxis.set_visible(False)

    plt.xlim(0,100)
    plt.ylim(0,100)
    plt.show()
    
    return pop

pop = pop_init(1000)
```
