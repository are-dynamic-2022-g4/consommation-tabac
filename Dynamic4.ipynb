```python
def create_pop(n):
    pop = []
    for i in range (n) :
        pop.append(person())
    return pop
```

```python
def interaction_proche(pop):
    proche = []
    for i in range(len(pop)):
        proche.append([])
    for i in range (len(pop)):
        x_val = [x[0] for x in pop]
        y_val = [x[1] for x in pop]
        dist = [np.sqrt((x_val[i] - x[0])**2 + (y_val[i] - x[1])**2) for x in pop]
        pgeo_i = [dist.index(j) for j in dist if j < 20]
        psoc_i = []
        for j in range(len(pgeo_i)):
            if i != j and j not in proche[i]:
                pi = pop[i]
                pj = pop[j]
                inter = random.random()
                if pi[5] == pj[5]:
                    if pi[3]-pj[3]<5:
                        if inter < 0.9:
                            proche[i].append(j)
                            proche[j].append(i)
                    elif pi[3]-pj[3]<15:
                        if inter < 0.5:
                            proche[i].append(j)
                            proche[j].append(i)
                    else:
                        if inter < 0.2:
                            proche[i].append(j)
                            proche[j].append(i)
                        
                else:
                    if pi[3]-pj[3]<5:
                        if inter < 0.3:
                            proche[i].append(j)
                            proche[j].append(i)
                    elif pi[3]-pj[3]<15:
                        if inter < 0.1:
                            proche[i].append(j)
                            proche[j].append(i)
                    else:
                        if inter < 0.03:
                            proche[i].append(j)
                            proche[j].append(i)


    i=0
    while i<len(proche):
        for j in proche[i]:
            change=[]
            if pop[j][-1]:
                for a in proche[i]:
                    if pop[a][-1]==False :
                        change.append(a)
            fume = random.random()
            if fume<0.3 :
                p = math.floor(len(change)*random.random())
                pop[p][-1]=True
                pop[p][2]='red'
        i=i+1

                 

        
```

```python
def diffu_pub(pop):
    radio = random.random()
    tv = random.random()
    reseau = random.random() 
    n = random.random()
    for i in range(len(pop)) :
        if pop[i][3]<41 :
            l = 8
        else :
            l = 12
        if pop[i][-2]<l:
            if radio<0.15:
                pop[i][-3]=pop[i][-3]+1
            if tv<0.25:
                pop[i][-3]=pop[i][-3]+1
            if pop[i][3]<60:
                if reseau<0.8:
                    pop[i][-3]=pop[i][-3]+1
            else:
                if reseau<0.2:
                    pop[i][-3]=pop[i][-3]+1
        if pop[i][-4]>0 :
            if pop[i][-1]==False:
                if pop[i][-3]>=l:
                    if n<0.25 :
                        pop[i][-1]=True
                        pop[i][2]='red'     
                elif pop[i][-3]>=l/2:
                    if n<0.011 :
                        pop[i][-1]=True
                        pop[i][2]='red'    
                else :
                    if n<0.005 :
                        pop[i][-1]=True
                        pop[i][2]='red'
```

```python
def diffu_lutte(pop):
    for i in range(len(pop)) :
        if pop[i][3]<41 :
            l = 6
        else :
            l = 10
        vu = random.random()
        if pop[i][-2]<l:
            if vu<0.25:
                pop[i][-2]=pop[i][-2]+1
        if pop[i][-4]<4 :
            if pop[i][-1]:
                n = random.random()
                if pop[i][-2]>=l/2:
                    if n<0.65 :
                        pop[i][-4]=pop[i][-4]+1
                        pop[i][-1]=False
                        pop[i][2]='white'
                        reprise = random.random()
                        if pop[i][-4]<4 :
                            if reprise<0.4 :
                                pop[i][-1]=True
                                pop[i][2]='red'
                else :
                    if n<0.3 :
                        pop[i][-4]=pop[i][-4]+1
                        pop[i][-1]=False
                        pop[i][2]='white'
                        reprise = random.random()
                        if pop[i][-4]<4 :
                            if reprise<0.7 :
                                pop[i][-1]=True
                                pop[i][2]='red'
```

```python
def affiche(pop, date = 0, titre = ''):
    
    fig = plt.figure(figsize = (6,6))
    
    x_val = [x[0] for x in pop]
    y_val = [x[1] for x in pop]
    color = [x[2] for x in pop]
            
    plt.scatter(x_val,y_val, c = color, edgecolors = 'black')

    ax = plt.gca()
    ax.axes.xaxis.set_visible(False)
    ax.axes.yaxis.set_visible(False)
    
    year = 2022 + date
    if titre == '':
        plt.title(str(year))
    else:
        plt.title(titre)

    plt.xlim(-1,101)
    plt.ylim(-1,101)
    plt.show()
    
    nbf = (sum(bool(x[-1]) for x in pop))*100/len(pop)
    print(str(nbf) + "% de fumeur")
```

```python
def year(pop):
    for i in range(len(pop)):
        pop[i][0] += random.randint(-5,5) 
        if pop[i][0]<0:
            pop[i][0] = random.randint(0,5)
        elif pop[i][0]>100:
            pop[i][0] = random.randint(95,100)
            
        pop[i][1] += random.randint(-5,5)
        if pop[i][1]<0:
            pop[i][1] = random.randint(0,5)
        elif pop[i][1]>100:
            pop[i][1] = random.randint(95,100)
            
    debute_age_sponta(pop)
    arret_reprise_sponta(pop)
    diffu_pub(pop)
    diffu_lutte(pop)
    interaction_proche(pop)
```

```python
def year_sans_pub(pop):
    for i in range(len(pop)):
        pop[i][0] += random.randint(-5,5) 
        if pop[i][0]<0:
            pop[i][0] = random.randint(0,5)
        elif pop[i][0]>100:
            pop[i][0] = random.randint(95,100)
            
        pop[i][1] += random.randint(-5,5)
        if pop[i][1]<0:
            pop[i][1] = random.randint(0,5)
        elif pop[i][1]>100:
            pop[i][1] = random.randint(95,100)
            
    debute_age_sponta(pop)
    arret_reprise_sponta(pop)
    diffu_lutte(pop)
    interaction_proche(pop)
```

```python
def year_sans_interaction(pop):
    for i in range(len(pop)):
        pop[i][0] += random.randint(-5,5) 
        if pop[i][0]<0:
            pop[i][0] = random.randint(0,5)
        elif pop[i][0]>100:
            pop[i][0] = random.randint(95,100)
            
        pop[i][1] += random.randint(-5,5)
        if pop[i][1]<0:
            pop[i][1] = random.randint(0,5)
        elif pop[i][1]>100:
            pop[i][1] = random.randint(95,100)
            
    debute_age_sponta(pop)
    arret_reprise_sponta(pop)
    diffu_pub(pop)
    diffu_lutte(pop)
```

```python
def affiche_simulation(nb_pop, nb_year):
    
    pop = pop_init(nb_pop)
    date = 0
    nbf = [sum(bool(x[10]) for x in pop)/len(pop)*100]
    affiche(pop,date)


    for i in range(1,nb_year+1):
        year(pop)
        date += 1
        nbf.append(sum(bool(x[-1]) for x in pop)/len(pop)*100)
        x_val = [x[0] for x in pop]
        y_val = [x[1] for x in pop]
        color = [x[2] for x in pop]
        
    affiche(pop,date)
    
    x = np.linspace(2022,2022+nb_year,nb_year+1)
    plt.plot(x,nbf)
    axes = plt.gca()
    axes.set_xlabel('Date')
    axes.set_ylabel('% de fumeur')
    plt.ylim(0,100)
    plt.xlim(2022,2022+nb_year)
    plt.show()
```
