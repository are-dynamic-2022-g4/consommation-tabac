```python
def create_pop(n):
    pop = []
    for i in range (n) :
        pop.append(person())
    return pop

create_pop(10)
```

```python
def pop_init(nb):
    pop = create_pop(nb)
    
    x_val = [x[0] for x in pop]
    y_val = [x[1] for x in pop]
    color = [x[2] for x in pop]
            
    plt.scatter(x_val,y_val, c = color, edgecolors = 'black')
    ax = plt.gca()
    ax.axes.xaxis.set_visible(False)
    ax.axes.yaxis.set_visible(False)
    plt.show()
    
    return pop

pop = pop_init(100)
```

```python
def proche(pop):
    proche = []
    for i in range (len(pop)):
        x_val = [x[0] for x in pop]
        y_val = [x[1] for x in pop]
        dist = [np.sqrt((x_val[i] - x[0])**2 + (y_val[i] - x[1])**2) for x in pop]
        p_i = [dist.index(j) for j in dist if j < 5]
        proche.append(p_i)
    print(proche)
        
```
